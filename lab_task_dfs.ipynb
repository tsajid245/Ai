```python
# Python program to print DFS traversal for complete graph
from collections import defaultdict
# This class represents a directed graph using adjacency
# list representation

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)

    def addEdge(self, u, v):
        self.graph[u].append(v)

    def DFSUtil(self, v, visited):
        visited[v] = True
        print(v, end=" ")
        for i in self.graph[v]:
            if not visited[i]:
                self.DFSUtil(i, visited)

    def DFS(self):
        V = max(self.graph.keys()) + 1   # ensure visited covers all vertices
        visited = [False] * V
        for i in range(V):
            if not visited[i]:
                self.DFSUtil(i, visited)

# Driver code
g = Graph()
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3, 3)

print("Following is Depth First Traversal")
g.DFS()
```

```python
from collections import defaultdict

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        self.weights = {}

    def addEdge(self, u, v, w):
        self.graph[u].append(v)
        self.weights[(u, v)] = w

    def DFSUtil(self, v, visited, goal, path, dist):
        visited[v] = True
        path.append(v)

        if v == goal:
            print("Path:", " -> ".join(path))
            print("Total Distance:", dist)
            return True

        for i in self.graph[v]:
            if not visited[i]:
                if self.DFSUtil(i, visited, goal, path, dist + self.weights[(v, i)]):
                    return True

        path.pop()
        return False

    def DFS(self, start, goal):
        # visited mein saare nodes dalna zaroori hai
        visited = {}
        for u in self.graph:
            visited[u] = False
            for v in self.graph[u]:
                if v not in visited:
                    visited[v] = False

        path = []
        if not self.DFSUtil(start, visited, goal, path, 0):
            print("No path found from", start, "to", goal)


# Driver code
g = Graph()

# Edges from the Romania map (subset sufficient for Arad → Bucharest)
g.addEdge("Arad", "Zerind", 75)
g.addEdge("Arad", "Sibiu", 140)
g.addEdge("Arad", "Timisoara", 118)

g.addEdge("Zerind", "Oradea", 71)
g.addEdge("Oradea", "Sibiu", 151)

g.addEdge("Sibiu", "Fagaras", 99)
g.addEdge("Sibiu", "Rimnicu Vilcea", 80)

g.addEdge("Fagaras", "Bucharest", 211)
g.addEdge("Rimnicu Vilcea", "Pitesti", 97)
g.addEdge("Rimnicu Vilcea", "Craiova", 146)

g.addEdge("Pitesti", "Bucharest", 101)
g.addEdge("Craiova", "Pitesti", 138)

g.addEdge("Timisoara", "Lugoj", 111)
g.addEdge("Lugoj", "Mehadia", 70)
g.addEdge("Mehadia", "Drobeta", 75)
g.addEdge("Drobeta", "Craiova", 120)

print("DFS Traversal from Arad to Bucharest:")
g.DFS("Arad", "Bucharest")

```

```python
from collections import defaultdict, deque

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        self.weights = {}

    def addEdge(self, u, v, w):
        self.graph[u].append(v)
        self.weights[(u, v)] = w

    # DFS
    def DFSUtil(self, v, visited, goal, path, dist):
        visited[v] = True
        path.append(v)

        if v == goal:
            print("DFS Path:", " -> ".join(path))
            print("DFS Distance:", dist)
            return True

        for i in self.graph[v]:
            if not visited[i]:
                if self.DFSUtil(i, visited, goal, path, dist + self.weights[(v, i)]):
                    return True

        path.pop()
        return False

    def DFS(self, start, goal):
        visited = {node: False for node in self.graph}
        for (u, v) in self.weights.keys():
            if v not in visited:
                visited[v] = False
        path = []
        self.DFSUtil(start, visited, goal, path, 0)

    # BFS
    def BFS(self, start, goal):
        visited = {node: False for node in self.graph}
        for (u, v) in self.weights.keys():
            if v not in visited:
                visited[v] = False

        queue = deque([(start, [start], 0)])
        visited[start] = True

        while queue:
            node, path, dist = queue.popleft()
            if node == goal:
                print("BFS Path:", " -> ".join(path))
                print("BFS Distance:", dist)
                return

            for i in self.graph[node]:
                if not visited[i]:
                    visited[i] = True
                    queue.append((i, path + [i], dist + self.weights[(node, i)]))


# Driver code
g = Graph()
# Add edges (subset sufficient for demo)
g.addEdge("Arad", "Zerind", 75)
g.addEdge("Arad", "Sibiu", 140)
g.addEdge("Arad", "Timisoara", 118)
g.addEdge("Zerind", "Oradea", 71)
g.addEdge("Oradea", "Sibiu", 151)
g.addEdge("Sibiu", "Fagaras", 99)
g.addEdge("Sibiu", "Rimnicu Vilcea", 80)
g.addEdge("Fagaras", "Bucharest", 211)
g.addEdge("Rimnicu Vilcea", "Pitesti", 97)
g.addEdge("Pitesti", "Bucharest", 101)
g.addEdge("Timisoara", "Lugoj", 111)
g.addEdge("Lugoj", "Mehadia", 70)
g.addEdge("Mehadia", "Drobeta", 75)
g.addEdge("Drobeta", "Craiova", 120)
g.addEdge("Craiova", "Pitesti", 138)

print("\nCase A: Arad → Bucharest (BFS < DFS)")
g.DFS("Arad", "Bucharest")
g.BFS("Arad", "Bucharest")

print("\nCase B: Timisoara → Bucharest (BFS > DFS)")
g.DFS("Timisoara", "Bucharest")
g.BFS("Timisoara", "Bucharest")
```
